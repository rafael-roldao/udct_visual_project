> PISA is a survey of students' skills and knowledge as they approach the end of compulsory education. It is not a conventional school test. Rather than examining how well students have learned the school curriculum, it looks at how well prepared they are for life beyond school.

> Around 510,000 students in 65 economies took part in the PISA 2012 assessment of reading, mathematics and science representing about 28 million 15-year-olds globally. Of those economies, 44 took part in an assessment of creative problem solving and 18 in an assessment of financial literacy.

### Math Analysis

```python
type(pisa_oecd.PV1MATH.iloc[0])
```

```python
pisa_oecd['PV1MATH'] = pisa_oecd['PV1MATH'].astype(float)
pisa_oecd['PV2MATH'] = pisa_oecd['PV2MATH'].astype(float)
pisa_oecd['PV3MATH'] = pisa_oecd['PV3MATH'].astype(float)
pisa_oecd['PV4MATH'] = pisa_oecd['PV4MATH'].astype(float)
pisa_oecd['PV5MATH'] = pisa_oecd['PV5MATH'].astype(float)
```

```python
pisa_oecd['PV1MATH'].describe()
```

```python
df_math = pisa_oecd.iloc[:, np.r_[:5, 500:505]]
df_math = df_math.groupby('CNT').mean()
df_math = df_math.reset_index()
df_math.head()
```

```python
g = sb.PairGrid(data=df_math, vars=['PV1MATH', 'PV2MATH', 'PV3MATH', 'PV4MATH', 'PV5MATH'])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
```

> Comments: I made the correlation charts above because the database includes 5 different scores for Math. However, given they are almost perfectly positive correlated, I will use only the first score ('PV1MATH') for my analysis.

```python
df_math_chart = df_math.sort_values('PV1MATH')

plt.figure(figsize=[8,10])
plt.barh('CNT', 'PV1MATH', data=df_math_chart)
plt.xlabel('PV1MATH');
```

> Comments: Raking of countries by Math score

> Comments: In order to facilitate the comparison between math, reading, and science scores by country, I ploted the charts side by side sorted by math scores.

## Multivariate Exploration

### Talk about some of the relationships you observed in this part of the investigation. Were there features that strengthened each other in terms of looking at your feature(s) of interest?

> In this section I will analyze the performance of students in math and reading areas by country. In addition, I will analize the relationship among math, reading and science scores.

### Were there any interesting or surprising interactions between features?

> More details below.

```python
df_scores_chart1 = df_scores_chart.sort_values('PV1MATH', ascending=False)
ss = sb.FacetGrid(data=df_scores_chart1, hue='CNT', height=8, palette='viridis');
ss.map(plt.scatter, 'PV1MATH', 'PV1READ')
ss.add_legend(title='Country');
```

> Comments: Ranking of countries by math score including reading score.

```python
plt.figure(figsize=[10,10])

plt.scatter(data=df_scores_chart, x='PV1MATH', y='PV1READ', c='PV1SCIE', cmap='viridis_r')
plt.colorbar(label='PV1SCIE');
plt.xlabel('PV1MATH')
plt.ylabel('PV1READ')
plt.title('Math vs. Read vs. Science (colorbar) scores');
```

> Comments: There is a high positive correlation among math, reading, and science scores, as expected.
