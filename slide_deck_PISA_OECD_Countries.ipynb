## Dataset Overview

> PISA is a survey of students' skills and knowledge as they approach the end of compulsory education. It is not a conventional school test. Rather than examining how well students have learned the school curriculum, it looks at how well prepared they are for life beyond school.

> Around 510,000 students in 65 economies took part in the PISA 2012 assessment of reading, mathematics and science representing about 28 million 15-year-olds globally. Of those economies, 44 took part in an assessment of creative problem solving and 18 in an assessment of financial literacy.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
pisa_oecd_original = pd.read_csv('pisa_oecd_2012.csv', sep=',', encoding='latin-1', error_bad_lines=False, index_col=0, dtype='unicode')
pisa_oecd = pisa_oecd_original.copy()
```

## Ranking of countries by math score

> The analysis below show the ranking of countries by math score. Considering the PISA assessment also inclues other learning areas (reading and science), the score of both areas was included side by side. 

> It is possible to see that some countries have a higher score in reading and science than in math. Indeed, the ranking would be different if sorted by other learning areas.

```python
pisa_oecd['PV1MATH'] = pisa_oecd['PV1MATH'].astype(float)
pisa_oecd['PV1READ'] = pisa_oecd['PV1READ'].astype(float)
pisa_oecd['PV1SCIE'] = pisa_oecd['PV1SCIE'].astype(float)

df_scores = pisa_oecd.iloc[:, np.r_[:5, 500, 540, 545]]
df_scores = df_scores.groupby('CNT').mean()
df_scores = df_scores.reset_index()
df_scores_chart = df_scores.sort_values('PV1MATH')

plt.figure(figsize=[12,10])

plt.subplot(1, 3, 1)
plt.barh('CNT', 'PV1MATH', data=df_scores_chart)
plt.xlabel('Math score');

plt.subplot(1, 3, 2)
plt.barh('CNT', 'PV1READ', data=df_scores_chart)
plt.xlabel('Reading score')
plt.yticks("")
plt.title('Ranking of countries by math score', fontsize=20);

plt.subplot(1, 3, 3)
plt.barh('CNT', 'PV1SCIE', data=df_scores_chart)
plt.xlabel('Science score')
plt.yticks("");
```

## Math lessons influence math scores?

> Unexpectedly, there is no evidence that math lessons positively influence math scores. Therefore, the countries should investigate the teaching methodologies to understand the reasons behind this conclusion.

```python
plt.figure(figsize=[20,40])
base_color = sb.color_palette()[0]

plt.subplot(4, 2, 1)
sb.boxplot(data=pisa_oecd, x='IC11Q01', y='PV1MATH', color=base_color)
plt.xlabel('Maths lessons - Draw graph', fontsize=15)
plt.ylabel('Math score', fontsize=12)
plt.xticks(rotation=12);

plt.subplot(4, 2, 2)
sb.boxplot(data=pisa_oecd, x='IC11Q02', y='PV1MATH', color=base_color)
plt.xlabel('Maths lessons - Calculation with numbers', fontsize=15)
plt.ylabel('Math score', fontsize=12)
plt.xticks(rotation=12);

plt.subplot(4, 2, 3)
sb.boxplot(data=pisa_oecd, x='IC11Q03', y='PV1MATH', color=base_color)
plt.xlabel('Maths lessons - Geometric figures', fontsize=15)
plt.ylabel('Math score', fontsize=12)
plt.xticks(rotation=12);

plt.subplot(4, 2, 4)
sb.boxplot(data=pisa_oecd, x='IC11Q04', y='PV1MATH', color=base_color)
plt.xlabel('Maths lessons - Spreadsheet', fontsize=15)
plt.ylabel('Math score', fontsize=12)
plt.xticks(rotation=12);

plt.subplot(4, 2, 5)
sb.boxplot(data=pisa_oecd, x='IC11Q05', y='PV1MATH', color=base_color)
plt.xlabel('Maths lessons - Algebra', fontsize=15)
plt.ylabel('Math score', fontsize=12)
plt.xticks(rotation=12);

plt.subplot(4, 2, 6)
sb.boxplot(data=pisa_oecd, x='IC11Q06', y='PV1MATH', color=base_color)
plt.xlabel('Maths lessons - Histograms', fontsize=15)
plt.ylabel('Math score', fontsize=12)
plt.xticks(rotation=12);
```

## Math vs. Reading scores by country

> As expected, countries with higher scores in math also have higher scores in reading. There is a high positive correlation between both variables.

```python
df_scores_chart1 = df_scores_chart.sort_values('PV1MATH', ascending=False)
ss = sb.FacetGrid(data=df_scores_chart1, hue='CNT', height=8, palette='viridis');
ss.map(plt.scatter, 'PV1MATH', 'PV1READ')
ss.add_legend(title='Country')
plt.xlabel('Math score', fontsize=12)
plt.ylabel('Reading score', fontsize=12)
plt.title('Math vs. Reading scores by country', fontsize=15);
```
